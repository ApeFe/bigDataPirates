```python
# Attribute equivalence blocking: block A, B  using birth_year
ab = em.AttrEquivalenceBlocker()
#C0 = ab.block_tables(A, B, 'Length_ft', 'Length_ft',
#                    l_output_attrs=['index', 'Length_ft'], 
#                    r_output_attrs=['index', 'Length_ft'])
```

```python
A['Length_ft'] = A['Length_ft'].astype(str)
B['Length_ft'] = B['Length_ft'].astype(str)
C0 = ab.block_tables(A, B, 'Length_ft', 'Length_ft',
                    l_output_attrs=['index', 'Length_ft', 'make_model', 'seller', 'location','year'], 
                    r_output_attrs=['index', 'Length_ft', 'Make_Model', 'seller', 'location','year'])
```

```python
len(C0)
```

```python
dbg = em.debug_blocker(C0, A, B)
len(dbg)
```

```python
sum(dbg['ltable_Length_ft'] == dbg['rtable_Length_ft'])
```

```python
A1 = A.set_index('index')
```

```python
B1 = B.set_index('index')
```

```python
a_col = []
b_col = []
for idx, row in dbg.iterrows():
    a_col.append(A1.ix[row['ltable_index']]['make_model'])
    b_col.append(B1.ix[row['rtable_index']]['Make_Model'])
dbg['ltable_make_model'] = a_col
dbg['rtable_make_model'] = b_col
```
